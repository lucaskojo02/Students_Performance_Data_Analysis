#### <u>Handling Outliers: Verification Before Removal</u>

Before deciding whether to remove or keep an outlier, you should verify it using **domain knowledge, data patterns, and external sources**. Hereâ€™s how:

##### 1. Check for Data Entry Errors
- Look at the **source of the data** (e.g., manual entry, database, or automated system).
- If it's a manual entry, check for **typos** (e.g., "101" instead of "100").
- If possible, **cross-check with original exam records**.

 **Action:** If it's a typo, **correct it**. If unsure, **consult the data provider**.


##### 2. Compare with Real-World Constraints
- **Example:** Exam scores are usually between **0 and 100**. A score of **101** is unrealistic unless extra credit was given.
- Check **official grading policies** to see if scores above 100 are possible.

 **Action:**  
- If scores above **100 are not possible**, **cap it at 100**.  
- If the grading system allows extra credit, **keep it**.


##### 3. Compare with Other Data Points
- Does the studentâ€™s **performance history** support an unusually high score?
- If all previous scores were around **60â€“80**, a sudden **101** is suspicious.
- Look at **related columns** (e.g., attendance, tutoring, hours studied).

 **Action:**  
- If the student has **low attendance and study hours but a perfect score**, **investigate further**.


##### 4. Check for Systematic Errors
- If **multiple students** have scores above 100, it could be a **grading system issue** rather than an individual error.
- If **only one student** has a 101, itâ€™s more likely an anomaly.

 **Action:**  
- If a **grading system issue** exists, **apply a global fix** (e.g., **rescaling scores**).


##### 5. Consult an Expert or Instructor
- If you have access to an **exam administrator or teacher**, ask if a **101 score** is realistic or a mistake.
- Sometimes, students get **extra credit points**, making scores above 100 possible.

 **Action:**  
- If the **expert confirms** the score, **keep it**.  
- If not, **adjust or remove it**.


#####  Conclusion
For the **101 in Exam_Score**, I recommend:

**Check the grading system**â€”if extra credit exists, **keep it**.  
If no extra credit is possible, **change it to 100**.  
If unsure, **flag it for review instead of removing it immediately**.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


# Initialize and train the model
linreg = LinearRegression()
linreg.fit(X_train, y_train)

# Predict on test data
y_pred_linreg = linreg.predict(X_test)

# Evaluate
mse_linreg = mean_squared_error(y_test, y_pred_linreg)
r2_linreg = r2_score(y_test, y_pred_linreg)

print(f"Linear Regression - MSE: {mse_linreg:.2f}, RÂ²: {r2_linreg:.2f}")
```

```python
# Start with numerical features with correlation > 0.1
corr_matrix = student_performance.corr(numeric_only=True)
strong_numeric_features = corr_matrix['Exam_Score'][corr_matrix['Exam_Score'].abs() > 0.1].index.tolist()
strong_numeric_features.remove('Exam_Score')  # remove target itself

# Add categorical features identified from EDA:
important_categoricals = ['Motivation_Level', 'Parental_Involvement', 'Parental_Education_Level',
                          'Teacher_Quality', 'Access_to_Resources', 'Family_Income', 'School_Type']

# Create a selected feature list
selected_features = strong_numeric_features + important_categoricals
print("Selected features:", selected_features)
```

#### Why Random Forest?
1. Handles non-linear relationships.
2. Captures feature interactions.
3. Handles both categorical (encoded) and numeric data easily.

#### Explanation:
1. Random Forest builds multiple decision trees and averages them to avoid overfitting.
2. We use 100 trees for stability.

```python
X_rf = student_performance.drop('Exam_Score', axis=1).copy()
y_rf = student_performance['Exam_Score']

# Encode categoricals if not done:
from sklearn.preprocessing import LabelEncoder
categorical_cols = X_rf.select_dtypes(include=['object']).columns.tolist()
le = LabelEncoder()
for col in categorical_cols:
    X_rf[col] = le.fit_transform(X_rf[col])

# Split into training and testing
from sklearn.model_selection import train_test_split
X_train_rf, X_test_rf, y_train_rf, y_test_rf = train_test_split(X_rf, y_rf, test_size=0.2, random_state=42)
```

#### Training the Random Forest Model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Initialize the model
rf_model = RandomForestRegressor(n_estimators=200, max_depth=15, random_state=42)
rf_model.fit(X_train_rf, y_train_rf)

# Predict
y_pred_rf = rf_model.predict(X_test_rf)

# Evaluate
mse_rf = mean_squared_error(y_test_rf, y_pred_rf)
r2_rf = r2_score(y_test_rf, y_pred_rf)

print(f"Random Forest Performance â€” MSE: {mse_rf:.2f}, RÂ²: {r2_rf:.2f}")
```

Why use Gradient Boosting?

Gradient Boosting builds trees sequentially, each correcting the errors of the previous.

Often more accurate than Random Forest if tuned well.

Handles non-linearity, feature interactions, and allows custom regularization.

```python
#Import and Prepare Data

import xgboost as xgb
from sklearn.metrics import mean_squared_error, r2_score

# Prepare data in DMatrix form (XGBoost's native format, optional but efficient)
dtrain = xgb.DMatrix(X_train_rf, label=y_train_rf)
dtest = xgb.DMatrix(X_test_rf, label=y_test_rf)
```

```python
#Train Gradient Boosting Model

params = {
    'objective': 'reg:squarederror',
    'max_depth': 6,
    'eta': 0.1,               # Learning rate
    'subsample': 0.8,         # Use 80% of data in each boosting round to prevent overfitting
    'colsample_bytree': 0.8,  # Use 80% of features per tree
    'seed': 42
}

# Train the model
xgb_model = xgb.train(params, dtrain, num_boost_round=200)
```

```python
#Predict and Evaluate

# Predict
y_pred_xgb = xgb_model.predict(dtest)

# Evaluate
mse_xgb = mean_squared_error(y_test_rf, y_pred_xgb)
r2_xgb = r2_score(y_test_rf, y_pred_xgb)

print(f"XGBoost - MSE: {mse_xgb:.2f}, RÂ²: {r2_xgb:.2f}")
```

```python
#Plot Feature Importances (XGBoost version)

xgb.plot_importance(xgb_model, max_num_features=10, importance_type='gain', height=0.5)
#plt.title('Top 10 Feature Importances (XGBoost)')
plt.show()
```
