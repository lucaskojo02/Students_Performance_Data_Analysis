# Milestone 1: Data Preprocessing and Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = "StudentPerformanceFactors2.csv"
student_performance = pd.read_csv(file_path)
student_performance
```

```python
#Basic information about the dataset
student_performance.info()
```

```python
# Getting summary statistics for numerical columns
summary_stats = student_performance.describe()
summary_stats
```

```python
#Checking for missing values
print(student_performance.isnull().sum())
```

```python
# Fill missing categorical values with mode (most frequent value)
student_performance["Distance_from_Home"] = student_performance["Distance_from_Home"].fillna(student_performance["Distance_from_Home"].mode().iloc[0])
student_performance["Parental_Education_Level"] = student_performance["Parental_Education_Level"].fillna(student_performance["Parental_Education_Level"].mode().iloc[0])
student_performance["Teacher_Quality"] = student_performance["Teacher_Quality"].fillna(student_performance["Teacher_Quality"].mode().iloc[0])

# Verify missing values are handled
student_performance.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

# Identify categorical columns
categorical_cols = student_performance.select_dtypes(include=['object']).columns
categorical_cols
```

```python
# Encoding categorical variables
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    student_performance[col] = le.fit_transform(student_performance[col])
    label_encoders[col] = le  # Store encoder for future reference
```

```python
# Display first few rows after encoding
student_performance.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Identify numerical columns (excluding the target variable 'Exam_Score')
numerical_cols = student_performance.select_dtypes(include=['int64']).columns.tolist()
numerical_cols.remove('Exam_Score')

numerical_cols
```

```python
# Apply standard scaling
scaler = StandardScaler()
student_performance[numerical_cols] = scaler.fit_transform(student_performance[numerical_cols])
```

```python
# Display first few rows after scaling
student_performance.head()
```

```python
# Check for duplicates
duplicate_count = student_performance.duplicated().sum()
duplicate_count
```

```python
from sklearn.ensemble import IsolationForest

# Define features (excluding target variable)
X = student_performance.drop(columns=["Exam_Score"])

# Apply Isolation Forest for anomaly detection
iso_forest = IsolationForest(contamination=0.05, random_state=42)  # Assume 5% anomalies
student_performance["Anomaly"] = iso_forest.fit_predict(X)

# Count anomalies (-1 indicates anomaly, 1 indicates normal)
anomaly_counts = student_performance["Anomaly"].value_counts()

# Extract anomalous data points
anomalies = student_performance[student_performance["Anomaly"] == -1]

# Display anomaly count and some anomalies
anomaly_counts
#print("\nSample Anomalies:\n", anomalies.head())
```

```python
anomalies.head()
```

```python
# Scatter plot: Study Hours vs. Exam Score (highlighting anomalies)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=student_performance, x="Hours_Studied", y="Exam_Score", hue="Anomaly", palette={1: "blue", -1: "red"})
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.title("Anomaly Detection in Student Performance (Isolation Forest)")
plt.legend(title="Anomaly", labels=["Normal", "Anomaly"])
plt.show()
```

* Blue (1) â†’ Normal students (perform as expected).

* Red (-1) â†’ Anomalous students (unexpected performance patterns).

* **Clusters of normal students** â†’ Most students follow a pattern (e.g., more study hours = better scores).

* **Anomalies (red points)** â†’ Some students deviate, like:

* **Studied a lot but scored low** (inefficient study habits, distractions, stress).

* **Studied little but scored high** (gifted students, prior knowledge).

```python
# Boxplot to visualize anomalies in Exam Score
plt.figure(figsize=(8, 5))
sns.boxplot(data=student_performance, x="Anomaly", y="Exam_Score", palette={1: "blue", -1: "red"},hue="Anomaly")
plt.xlabel("Anomaly (1: Normal, -1: Anomaly)")
plt.ylabel("Exam Score")
plt.title("Boxplot of Anomalies in Exam Score")
plt.show()
```

* **Blue** = Normal students.

* **Red** = Anomalous students.

* **Median line (middle of the box)** â†’ Typical exam scores for normal and anomalous students.

* **Box spread (IQR - Interquartile Range)** â†’ Shows score distribution.

* **Outliers (dots outside whiskers)** â†’ Extreme performances.


### Insights:

* If anomalous studentsâ€™ scores have a wider range, they have inconsistent or unexpected performance.

* If normal students have tightly packed scores, their performance is more predictable.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
import pandas as pd

# Prepare the data
X = student_performance[features].dropna()  # Make sure 'features' is defined as your list of numeric features
X = add_constant(X)

# Create VIF dataframe
vif_data = pd.DataFrame()
vif_data['feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# Display VIF
print(vif_data)
```
