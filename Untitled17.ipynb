# Milestone 1: Data Preprocessing and Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = "StudentPerformanceFactors2.csv"
student_performance = pd.read_csv(file_path)
student_performance
```

```python
#Basic information about the dataset
student_performance.info()
```

```python
# Getting summary statistics for numerical columns
summary_stats = student_performance.describe()
summary_stats
```

```python
#Checking for missing values
print(student_performance.isnull().sum())
```

```python
# Fill missing categorical values with mode (most frequent value)
student_performance["Distance_from_Home"] = student_performance["Distance_from_Home"].fillna(student_performance["Distance_from_Home"].mode().iloc[0])
student_performance["Parental_Education_Level"] = student_performance["Parental_Education_Level"].fillna(student_performance["Parental_Education_Level"].mode().iloc[0])
student_performance["Teacher_Quality"] = student_performance["Teacher_Quality"].fillna(student_performance["Teacher_Quality"].mode().iloc[0])

# Verify missing values are handled
student_performance.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

# Identify categorical columns
categorical_cols = student_performance.select_dtypes(include=['object']).columns
categorical_cols
```

```python
# Encoding categorical variables
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    student_performance[col] = le.fit_transform(student_performance[col])
    label_encoders[col] = le  # Store encoder for future reference
```

```python
# Display first few rows after encoding
student_performance.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Identify numerical columns (excluding the target variable 'Exam_Score')
numerical_cols = student_performance.select_dtypes(include=['int64']).columns.tolist()
numerical_cols.remove('Exam_Score')

numerical_cols
```

```python
# Apply standard scaling
scaler = StandardScaler()
student_performance[numerical_cols] = scaler.fit_transform(student_performance[numerical_cols])
```

```python
# Display first few rows after scaling
student_performance.head()
```

```python
# Check for duplicates
duplicate_count = student_performance.duplicated().sum()
duplicate_count
```

## Creating visualizations

```python
# Boxplot for detecting outliers in Exam Scores

# Set plot style
sns.set_style("whitegrid")

plt.figure(figsize=(6, 4))
sns.boxplot(x=student_performance["Exam_Score"], color='red')
plt.title("Boxplot of Exam Scores")
plt.xlabel("Exam Score")
plt.show()
```

```python
# Histogram of exam scores
plt.figure(figsize=(8, 5))
sns.histplot(student_performance['Exam_Score'], bins=20, kde=True)
plt.title("Distribution of Exam Scores")
plt.xlabel("Exam_Score")
plt.ylabel("Number of Students")
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(15, 10))
corr_matrix = student_performance.corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm",fmt=".2f", linewidths=0.5, annot_kws={"size": 10},cbar=True)
plt.title("Correlation Matrix of Features")
plt.show()
```

## Key Correlations with Exam Score:

```python
from sklearn.ensemble import IsolationForest

# Define features (excluding target variable)
X = student_performance.drop(columns=["Exam_Score"])

# Apply Isolation Forest for anomaly detection
iso_forest = IsolationForest(contamination=0.05, random_state=42)  # Assume 5% anomalies
student_performance["Anomaly"] = iso_forest.fit_predict(X)

# Count anomalies (-1 indicates anomaly, 1 indicates normal)
anomaly_counts = student_performance["Anomaly"].value_counts()

# Extract anomalous data points
anomalies = student_performance[student_performance["Anomaly"] == -1]

# Display anomaly count and some anomalies
anomaly_counts
#print("\nSample Anomalies:\n", anomalies.head())
```
